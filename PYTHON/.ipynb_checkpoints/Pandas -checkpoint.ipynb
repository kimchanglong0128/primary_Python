```python
daterng = pd.Series(pd.date_range('2017', periods=9, freq='Q'))   ### freq = 'Q'         B,(M,Q,A),S	分别代表了工作日，（以月为频率，以季度为频率，以年为频率），最接近月初的那一天
print(daterng)
```

```python
daterng.dt.day_name()    ### 从日期判断所处的星期数
```

```python
print(daterng[daterng.dt.quarter < 5])  ### Series.dt.quarter：从日期判断所处季节
```

```python
print(daterng[daterng.dt.is_year_end])   ### Series.dt.is_year_end：从日期判断是否处在年
```
