```python
import pandas as pd
import numpy as np

DF = pd.read_excel(r"C:\Users\jinchanglong\Desktop\10.LAMI.xlsx")
df = pd.read_excel(r"C:\Users\jinchanglong\Desktop\16.STK_l.xlsx")

display(DF)
print("----------------------------------------------------------------------------------------------------------------------------------------------------")
display(df)
```

###### 左右连接键名一样

```python
df1 = pd.merge(df,DF)
display(df1)
```

```python
df2 = pd.merge(df,DF , on = 'CELL_ID' , how = 'inner')    ### 以df，DF 中相同的列名CELL_ID进行连接，默认how = 'inner''
display(df2)
```

###### 全连接，取并集

```python
df3 = pd.merge(df,DF,how = 'outer')
display(df3)
```

###### 左连接，左边取全部，右边取部分，没有值则用NaN填充

```python
df4 = pd.merge(df,DF,how = 'left')
display(df4)
```

######  右连接，右边取全部，左边取部分，没有值则用NaN填充

```python
df5 = pd.merge(DF,df,how = 'right')
display(df5)
```

###### 左右连接的键名不一样

###### 内连接

```python
df6 = pd.merge(df,DF,left_on = 'CELL_ID',right_on = 'LOT_ID', how = 'inner')
display(df6)
```

###### 全连接

```python
df7 = pd.merge(df,DF,left_on = 'CELL_ID',right_on = 'LOT_ID',how = 'outer')
display(df7)
```

###### 左连接

```python
df8 = pd.merge(df,DF,left_on='CELL_ID',right_on='LOT_ID',how='left') 
display(df8)
```

```python
df9 = pd.merge(df,DF,left_on='CELL_ID',right_on='LOT_ID',how='right') 
display(df9)
```

###### 索引作为连接键

```python
df10 = pd.merge(df,DF,left_index = True ,right_index = True)
display(df10)
```
