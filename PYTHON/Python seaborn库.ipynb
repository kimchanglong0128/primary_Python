### set_style( )  set( )

```python
import matplotlib.pyplot as plt  
import numpy as np
import seaborn as sns  
sns.set_style("ticks")  
plt.plot(np.arange(10))  
plt.show()  
```

```python
import seaborn as sns  
import matplotlib.pyplot as plt  
sns.set(style = "white", palette = "muted", color_codes = True)     #set( )设置主题，调色板更常用  
plt.plot(np.arange(10))  
plt.show()  
```

### distplot( )  kdeplot( )

```python
import pandas as pd
import matplotlib.pyplot as plt  
import seaborn as sns  
df_iris = pd.read_csv('iris.csv')  
fig, axes = plt.subplots(1,2)  
sns.distplot(df_iris['petal_length'], ax = axes[0], kde = True, rug = True)        # kde 密度曲线  rug 边际毛毯  
sns.kdeplot(df_iris['petal_length'], ax = axes[1], shade=True)                     # shade  阴影                         
plt.show()  
```

```python
import numpy as np  
import seaborn as sns  
import matplotlib.pyplot as plt  
sns.set( palette="muted", color_codes=True)  
rs = np.random.RandomState(10)  
d = rs.normal(size=100)  
f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)  
sns.distplot(d, kde=False, color="b", ax=axes[0, 0])  
sns.distplot(d, hist=False, rug=True, color="r", ax=axes[0, 1])  
sns.distplot(d, hist=False, color="g", kde_kws={"shade": True}, ax=axes[1, 0])  
sns.distplot(d, color="m", ax=axes[1, 1])  
plt.show()  
```

### 箱型图

```python
import matplotlib.pyplot as plt  
import seaborn as sns  
df_iris = pd.read_csv('iris.csv')  
sns.boxplot(x = df_iris['class'],y = df_iris['sepal_width'])  
plt.show()  
```

### 联合分布jointplot()

```python
tips = pd.read_csv('Iris.csv')   #右上角显示相关系数  
sns.jointplot("sepal_length", "petal_length", tips)  
plt.show()  
```
