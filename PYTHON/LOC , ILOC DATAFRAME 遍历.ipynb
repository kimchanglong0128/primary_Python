通过行标签进行筛选，会用到loc

```python
import pandas as pd
import numpy as np
df = pd.read_excel(r"C:/Users/jinchanglong/Desktop/text1.xlsx")
df1 = pd.read_excel(r"C:/Users/jinchanglong/Desktop/text1.xlsx",index_col = 0)
df1
```

### LOC

#### LOC 单行单列筛选

```python
df.loc[1,'性别']    ### 当前边为索引时，不用加引号
```

```python
df1.loc['鳄鱼','性别']   ### 若是索引为字符串时，要加引号
```

#### LOC 某行多列筛选

```python
df.loc[1,['性别','体重']]
```

```python
df1.loc['鳄鱼',['籍贯','年龄','性别']]
```

#### 获取某行数据

```python
df.loc[1,:]
```

```python
df1.loc['鳄鱼',:]
```

#### 选择间隔的多行多列

```python
df.loc[[1,2,3,4,5,6,7],['年龄','籍贯']]
```

```python
df1.loc[['鳄鱼','骨龙','山猫','乌鸦','兔子','麻雀'],['年龄','籍贯']]
```

#### 选择连续的多行多列 

```python
df.loc[1:6,'性别':'体重']
```

```python
df1.loc['丑八怪':'麻雀','性别':'体重']
```

### ILOC

行位置签索引进行筛选

#### 获取单行数据

```python
df.iloc[0]
```

```python
df1.iloc[0]
```

#### 获取连续多行数据

```python
df.iloc[0:6]
```

```python
df1.iloc[0:6]
```

#### 获取间隔多行

```python
df.iloc[[0,1,2,3,4,5,6],:]
```

```python
df.iloc[: ,:]
```

### DataFrame 遍历

```python
df.iterrows()
```

```python
df.iteritems()
```

```python
for index , row_data in df.iteritems():       
    print(index ,row_data) 
    
    print(index)
    print(row_data)
```

```python
for index , row_data  in df.iterrows():   
    print(index ,row_data) 

    
    print(index)
    print(row_data)
```

```python
for columns , col_data in df.iteritems(): 
    print(columns , col_data)
```

```python
for columns , col_data in df.iterrows(): 
    print(columns , col_data)    
```
