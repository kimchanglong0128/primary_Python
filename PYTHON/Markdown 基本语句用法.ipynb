#### base64的图片编码

##### 将图片转化为base64字符串

````python
import base64
f = open('1.jpg','rb') #二进制方式打开图文件
ls_f = base64.b64encode(f.read()) #读取文件内容，转换为base64编码
f.close()
print(ls_f)
````

##### base64字符串转化为图片

````python
import base64
imgdata=base64.b64decode(ls_f)
file=open('2.jpg','wb')
file.write(imgdata)
file.close()
````
